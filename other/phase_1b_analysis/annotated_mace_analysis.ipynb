```python
moralitys_list = []
concerns_list = []
emotions_list = []
for message in messages:
    morality_dict = {}
    concern_dict = {}
    emotion_dict = {}
    annotation = message["annotations"]
    for morality in morality_list:
        morality_dict[morality] = annotation[morality]
    for concern in concern_list:
        concern_dict[concern] = annotation[concern]
    for emotion in emotion_list:
        emotion_dict[emotion] = annotation[emotion]
    moralitys_list.append(morality_dict)
    concerns_list.append(concern_dict)
    emotions_list.append(emotion_dict)
```

```python
def cal_confidence(df, target_list, type):
    # df = df[df["type"] == type]
    date_list =df["date"].tolist()
    date_set = list(set(date_list))
    date_dict = {}
    target_dict = {}

    for date in date_set:
        date_dict[date] = 0

    for date in date_set:
        annotation_dict = {}
        for target in target_list:
            annotation_dict[target] = 0.0
        target_dict[date] = annotation_dict

    for row in tqdm(df.itertuples(), total=len(df)):
        tweet_date = row[1]
        tweet_type = row[2]
        annotation_dict = row[3]

        date_dict[tweet_date] += 1

        for target in target_list:
            target_dict[tweet_date][target] += annotation_dict[target]

    for date in date_set:
        result_dict = {}
        for target in target_list:
            result_dict[target] = target_dict[date][target] / date_dict[date]
        target_dict[date] = result_dict

    result_date_list = list(target_dict.keys())
    result_annotation_list = list(target_dict.values())
    result_df = pd.DataFrame(result_annotation_list)
    result_df["date"] = result_date_list
    return result_df
```
